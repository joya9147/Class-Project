```python
import pandas as pd
import numpy as np
```

```python
kills_df = pd.read_csv('../raw_data/kills.csv')
match_df = pd.read_csv('../raw_data/matchinfo.csv')
```

```python
match_df.head()
```

Replacing player names with their roles.

```python
blue_tag = match_df['blueTeamTag'].astype(str)
red_tag = match_df['redTeamTag'].astype(str)

roles = match_df.drop(match_df.columns[0:9],axis = 1)
roles.head()
```

```python
for i in roles.columns:
    if 'Champ' in i:
        roles = roles.drop(i, axis = 1)
```

```python
roles.head()
```

```python
def PlayerToRole(i):
    x = kills_df[kills_df['Address'] == roles['Address'][i]]
    x = x.replace(dict(zip(list(blue_tag[i]+ ' ' + roles.loc[i][:5]) + list(red_tag[i]+ ' ' + roles.loc[i][5:10]),
              list(roles.columns[0:10]))))
    return x
```

```python
kills = pd.DataFrame(columns = kills_df.columns)

for i in np.arange(0,roles.shape[0]):
    x = PlayerToRole(i)
    kills = pd.concat([kills,x.copy()],axis=0)
```

```python
kills
```

Seperating red and blue team stats for each roles. The values will be the total kills and assists for a role within 15 minutes.

```python
blue = pd.DataFrame()
bkills = kills[kills['Team'] == 'bKills']
for i in kills['Address'].unique():
    x = bkills[bkills['Address'] == i]
    z = pd.DataFrame([
        (i,
        x[['Killer']].eq('blueMiddle').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('blueMiddle').sum().sum(),
        x[['Killer']].eq('blueTop').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('blueTop').sum().sum(),
        x[['Killer']].eq('blueADC').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('blueADC').sum().sum(),
        x[['Killer']].eq('blueSupport').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('blueSupport').sum().sum(),
        x[['Killer']].eq('blueJungle').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('blueJungle').sum().sum())])
    blue = pd.concat([blue,z],axis = 0)
```

```python
blue.columns = [['Info','Blue','Blue','Blue','Blue','Blue','Blue','Blue','Blue','Blue','Blue'],
                ['MatchID','Middle','Middle','Top','Top','ADC','ADC','Support','Support','Jungle','Jungle'],
               ['Address','Kills','Assist','Kills','Assist','Kills','Assist','Kills','Assist','Kills','Assist']]
```

```python
red = pd.DataFrame()
for i in kills['Address'].unique():
    x = kills[kills['Address'] == i]
    x = x[x['Team'] != 'bKills']
    z = pd.DataFrame([
        (i,
        x[['Killer']].eq('redMiddle').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('redMiddle').sum().sum(),
        x[['Killer']].eq('redTop').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('redTop').sum().sum(),
        x[['Killer']].eq('redADC').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('redADC').sum().sum(),
        x[['Killer']].eq('redSupport').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('redSupport').sum().sum(),
        x[['Killer']].eq('redJungle').sum()[0],
        x[['Assist_1','Assist_2','Assist_3','Assist_4']].eq('redJungle').sum().sum())])
    red = pd.concat([red,z],axis = 0)
```

```python
red.columns = [['Info','Red','Red','Red','Red','Red','Red','Red','Red','Red','Red'],
                ['MatchID','Middle','Middle','Top','Top','ADC','ADC','Support','Support','Jungle','Jungle'],
               ['Address','Kills','Assist','Kills','Assist','Kills','Assist','Kills','Assist','Kills','Assist']]
```

```python
kills = pd.merge(blue,red)
```

```python
kills.to_csv('../clean_data/kills_cleaned.csv')
```
