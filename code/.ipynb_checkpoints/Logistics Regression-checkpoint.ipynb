```python
pipe = Pipeline(steps=[('encoder', columns), ('model', model)])
```

```python
pipe.fit(X_train, y_train)
```

```python
#y_pred = pipe.predict(X_val)
#util.print_accuracies(pipe, X_train, y_train, X_val, y_val)
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=20220914)

scores = cross_val_score(pipe, X_train, y_train, cv=skf)
```

```python
param_grid = {
    'model__penalty': ['l2', 'l1']
#    'model__C': np.logspace(-4, 0, 5)
    
}

grid_search = GridSearchCV(pipe, param_grid=param_grid)
```

```python
grid_search.fit(X_train, y_train)
```

```python
params = grid_search.cv_results_['params']
scores = grid_search.cv_results_['mean_test_score']
```

```python
rows = [dict(param_dict, score=score) for param_dict, score in zip(params, scores)]
```

```python
results_df = pd.DataFrame(rows)
results_df.columns = ['Reg. Type', 'Accuracy']
#results_df.columns = ['Reg. Strength', 'Reg. Type', 'Accuracy']
results_df
```
